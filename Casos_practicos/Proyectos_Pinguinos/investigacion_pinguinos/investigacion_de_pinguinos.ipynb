#### *se genera el diagrama con scatterplot usando los datos de species para diferenciar la nube de puntos*
#### *se genera la matriz con pairplot usando todas las variables de datos*

```python
fig,ax = plt.subplots()
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=df_penguins, hue='species')
fig.text(0.75, 0.01, f'Creado el {ahora}')

g = sns.pairplot(data=df_penguins, hue='species', height=2.5)
g.figure.text(0.84, 0.01, f'creado el {ahora}')

plt.grid(True)

plt.show()
```
